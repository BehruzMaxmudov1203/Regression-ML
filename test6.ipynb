```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
housing = df[df.district == "Чиланзарский"]
housing.head()
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(housing,test_size=0.10,random_state=42)
```

```python
housing.corr(numeric_only=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()

x_train = np.asanyarray(train_set[['size','max_levels']])
y_train  = np.asanyarray(train_set['price'])
```

```python
model.fit(x_train,y_train)
```

```python
print('Coefficients: ',model.coef_)
print('theta0: ',model.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df[['rooms','size','level','max_levels','price']])
plt.show()
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
housing = df[df.district == "Чиланзарский"]
housing.head()
```

```python
#X = np.asanyarray(housing[['rooms','size','level','max_levels']])
X = np.asanyarray(housing[['size','max_levels']])
Y = np.asanyarray(housing['price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.10,random_state=42)
```

```python
from sklearn import  linear_model
LR_model = linear_model.LinearRegression()

LR_model.fit(x_train,y_train)

print(f"{LR_model.coef_ }")
print(f"{LR_model.intercept_}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.pairplot(df[['rooms','size','level','max_levels','price']])
plt.show()
```

# Modelni baholaymiz

```python
y_predict = LR_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

MAE = mean_absolute_error(y_test,y_predict)
RMSE = np.sqrt(mean_squared_error(y_test,y_predict))

print(f"{MAE = }")
print(f"{RMSE = }")
```
