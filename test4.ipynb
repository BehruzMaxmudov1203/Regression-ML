```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Dataset
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)

features = ['size', 'rooms', 'level', 'max_levels']
districts = df['district'].unique()

# Natijalarni saqlash uchun DataFrame
results = pd.DataFrame(columns=['District', 'Feature', 'Theta0', 'Theta1', 'MAE', 'RMSE'])

def run_simple_lr(housing, district, feature):
    X = housing[[feature]]
    Y = housing['price']
    
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    
    model = LinearRegression()
    model.fit(x_train, y_train)
    
    theta1 = model.coef_[0]
    theta0 = model.intercept_

    print(f"{theta1 = }")
    print(f"{theta0 = }")
    
    y_pred = model.predict(x_test)
    
    MAE = mean_absolute_error(y_test, y_pred)
    RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Natijalarni DataFrame ga qo‘shish
    global results
    results = pd.concat([results, pd.DataFrame({
        'District': [district],
        'Feature': [feature],
        'Theta0': [theta0],
        'Theta1': [theta1],
        'MAE': [MAE],
        'RMSE': [RMSE]
    })], ignore_index=True)
    
    # Vizualizatsiya
    plt.figure(figsize=(6,4))
    plt.scatter(x_test, y_test, color='blue', alpha=0.6, label='Actual Price')
    plt.plot(x_test, y_pred, color='red', label='Predicted Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(f"{district}: {feature} vs Price")
    plt.legend()
    plt.show()
    
# Barcha tuman va featurelar uchun regressionni bajarish
for district in districts:
    housing = df[df['district'] == district]
    for feature in features:
        run_simple_lr(housing, district, feature)

# Natijalarni ko‘rsatish
print(results)

# CSV sifatida saqlash
results.to_csv('district_feature_lr_results.csv', index=False)
```
